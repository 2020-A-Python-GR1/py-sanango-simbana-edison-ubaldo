```python
# 7 )

df = pd.DataFrame(np.random.randint(0,60,size=(10, 6)), columns=['C1', 'C2','C3','C4','C5','C6'])

df_c1_a = df.sort_values(by='C1', ascending = True)
print(df_c1_a) 
df_c1_d = df.sort_values(by='C1', ascending = False)
print(df_c1_d)
print("********************")
df_c2_a = df.sort_values(by='C2', ascending = True)
print(df_c2_a) 
df_c2_d = df.sort_values(by='C2', ascending = False)
print(df_c2_d)
print("********************")
df_c3_a = df.sort_values(by='C3', ascending = True)
print(df_c3_a) 
df_c3_d = df.sort_values(by='C3', ascending = False)
print(df_c3_d)
print("********************")
df_c4_a = df.sort_values(by='C4', ascending = True)
print(df_c4_a) 
df_c4_d = df.sort_values(by='C4', ascending = False)
print(df_c4_d)
print("********************")
df_c5_a = df.sort_values(by='C5', ascending = True)
print(df_c5_a) 
df_c5_d = df.sort_values(by='C5', ascending = False)
print(df_c5_d)
print("********************")
df_c6_a = df.sort_values(by='C6', ascending = True)
print(df_c6_a) 
df_c6_d = df.sort_values(by='C6', ascending = False)
print(df_c6_d)
print("********************")
```

```python
# 12 ) 

dict_ = {
    'C1': ['a1','b1','c1','d1','e1','f1','g1','h1','i1','j1'],
    'C2': ['a2','b2','c2','d2','e2','f2','g2','h2','i2','j2'],
    'C3': ['a3','b3','c3','d3','e3','f3','g3','h3','i3','j3'],
    'C4': ['a4','b4','c4','d4','e4','f4','g4','h4','i4','j4'],
    'C5': ['a5','b5','c5','d5','e5','f5','g5','h5','i5','j5'],
    'C6': ['a6','b6','c6','d6','e6','f6','g6','h6','i6','j6']
}

df = pd.DataFrame(dict_)
df_final = pd.DataFrame()
df_final['C1C2'] = df['C1'] + df['C2']
df_final['C3C4'] = df['C3'] + df['C4']
df_final['C5C6'] = df['C5'] + df['C6']
df_final
```
